# EDA Questions

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('../data/Computersales.csv')
```

```python
df.head()
```

### Q.1) Plot countplot to show the frequency of each Product Type?

```python
sns.countplot(data=df,x="Product Type")
plt.title("Frequency of Product Type")
plt.xlabel("Product Type")
plt.ylabel("Frequency")
plt.show()
```

### Q.2) Plot a barplot to show yearly profit?

```python
sns.barplot(data=df,x="Year",y="Profit")
plt.title("Yearly Profit")
plt.xlabel("Year")
plt.ylabel("Profit")
plt.show()
```

### Q.3) Plot CountPlot to show the frequency of Lead Column?

```python
sns.countplot(data=df,x="Lead")
plt.title("Frequency of Lead Column")
plt.xlabel("Lead")
plt.ylabel("Frequency")
plt.show()
```

### Q.4) Plot the average profit for each month? 

```python
avg=df.groupby('Month')["Profit"].mean().reset_index()
sns.lineplot(data=avg,x="Month",y="Profit")
plt.title("Average profit for each month")
plt.show()
```

### Q.5) Create a box plot to visualize the distribution & outliers of "Profit" column.


```python
sns.boxplot(data=df,x="Profit")
plt.title("Distribution of Profit")
plt.xlabel("Profit")
plt.show()
```

### Q.6) Plot a pie chart of the product types column?

```python
plt.pie(df['Product Type'].value_counts(),autopct='%1.1f%%')
plt.legend(df['Product Type'].unique())
plt.show()
```

### Q.7) Create a bar chart of the total sales by state?

```python
grp = df['Sale Price'].groupby(df['State'])
mean_sale_price_by_state = grp.mean().reset_index()
sns.barplot(data=mean_sale_price_by_state, x='State', y='Sale Price')
plt.show()
```

### Q.8) How can you create a heatmap using Seaborn that shows the correlation between variables in a Pandas DataFrame?


```python
numeric_columns = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_columns.corr() 
sns.heatmap(correlation_matrix, annot=True, cmap='RdYlGn')
plt.show()
```

### Q.9) Load the iris dataset from Seaborn and create a line plot showing the changes in sepal length over time for all three species.

```python
sns.scatterplot(data=iris,x="sepal_length",y="sepal_width",hue='species')
plt.show()
```

### Q.12)Create a box plot of sepal length for each species, with the box color set to a different color for each.


```python
sns.boxplot(data=iris,x="sepal_length",y='species',hue='species')
plt.show()
```
